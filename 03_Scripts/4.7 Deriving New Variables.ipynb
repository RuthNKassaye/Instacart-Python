###  A user-defined function: sortting products into price ranges
1. If the item’s price is lower than or equal to $5, it will be labeled a “low-range product.”

2. If the item’s price is above USD 5 but lower than or equal to $15, it will be labeled a a “mid-range product.”
3. If the item’s price is above USD 15, it will be labeled a “high-range product.”

```python
# Creating a subset of the ords_prods_merge dataframe as the dataframe is big. 
# Creating subset including everything starting from the beginning of the datafram up to 1 million

df= ords_prods_merge[:1000000]
```

```python
df.head()
```

```python
def price_label(row):
    if row['prices'] <=5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

### Designating a location for the 'price_lable' column

```python
# Creating a new column within the df dataframe called “price_range” . axis = 1 tells panda to apply the function to all rows within the dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Counting the records in the  price_range column
df['price_range'].value_counts(dropna = False)
```

## 2. If-Statements with the loc() Function

- loc() function locates a particular column in the dataframe it’s been assigned to.

```python
# Breaking down the code below called 
# if = df.loc[df['prices'] > 15, then = 'price_range_loc'] = 'High-range product'

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Breaking down the code below called 
#if = df.loc[(df['prices'] <= 15) & (df['prices'] > 5), then = 'price_range_loc'] = 'Mid-range product'

df.loc[(df['prices'] <=15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
# df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
#df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'

df.loc[df['prices'] <=5, 'price_range_loc'] = 'Low-range product'
```

```python
# Counting the records in the  price_range_loc column
df['price_range_loc'].value_counts(dropna = False)
```

### Creating columns for 'Busiest day', 'Least busy', 'Regulary busy' using for-loop

```python
# Creating a dataframe using for-loop based in business categories and saving the categories under result
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

### Q4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

#### First step: know which hours have the most orders, fewest orders and average ordres using frequncy function

```python
# Printing frequency of busiest order_hours_of_day column to see what are the most orders, fewest orders and average orders
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

#### Second step: Creating label list - most orders, fewest orders and average ordres from the order_hour_of_day column
 - 10 o'clock would be the hour with the highest orders and 3 o'clock is the hour with the lowest order. Else, would be labeled under average orders

```python
# Creating labels “Most orders,” “Average orders,” and “Fewest orders.”

Busiest_period_of_day = []

for value in ords_prods_merge ['order_hour_of_day']: 
  if value == 10 or value == 11 or value == 14 or value ==15 or value ==13 or value ==12 or value ==16 or value == 9:
    Busiest_period_of_day.append('Most orders')
  elif value == 3 or value ==4 or value ==2 or value ==5 or value ==1 or value ==0 or value ==6 or value ==23:
    Busiest_period_of_day.append('Fewest orders')
  else: 
    Busiest_period_of_day.append('Average orders')
```

```python
# Calling the Busiest_period_of_day label list created 
Busiest_period_of_day
```

#### Third Step: Adding the Busiest_period_of_day label list I created above in the ords_prods_merge dataframe

```python
# Creating a column busiest_period_of_day and assign Busiest_period_of_day label list value to it, see above
ords_prods_merge['busiest_period_of_day'] = Busiest_period_of_day 
```
