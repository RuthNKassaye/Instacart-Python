## Transposing data
- Transposing means turning my dataframe’s rows into columns and vise versa.
- **Why trasposing?**: sometimes dataframes could be wide. See for instance df_dep dataframe imported above.There’s only 1 row and 22 columns, making the dataframe incredibly wide and incredibly short. This is what “wide format” looks like.

## Step 1:  Trasposing the imported dataframe above (**df_dep**) and assigning a new dataframe to the trasposed dataframe(**df_dep_t**)

```python
# Trasposing df_dep dataframe and assigning for the trasposed dataframe a new dataframe, df_dept_t
df_dep_t = df_dep.T
```

```python
# Printing the entire rows of the df_dept_t dataframe
df_dep_t
```

## Step 2: Creating an index
- This function is the first step to remove the 0 header. See above the trasposed dataframe. We don't want to be presented like this.

```python
# adding and index in the header. Note that I have two rows of headers. So I need to have only one header
df_dep_t.reset_index()
```

## Step 3: Creating a new header 
- This is the header I want to have in my **df_dep_t** dataframe
- **iloc[0]** means I am telling panda to select index 0 or the first row as my new header. In the above step I had two headers. So I am choosing here which of the headers to be my real header

```python
# Creating a new header (new_header).   
new_header = df_dep_t.iloc[0]
```

```python
new_header
```

## Step 4: Removing the first row from the df_dep_t dataframe and copy the dataframe into a new daraframe
- Now that I have selected index 0 to be my new_header. I will copy every rows below my new heade starting from **index 0** or **first row** into a new dataframe **(df_de_t_new)**

```python
# Tellign Python to copy everything starting from first row/index 0 
df_dep_t_new = df_dep_t[1:]
```

```python
# Printing out the df_dep_t_new dataframe. Note that this looks exactly like the old dataframe only without the first row. I should now set my new_header variable as the new header for my df_dep_t_new dataframe
df_dep_t_new
```

## Step 5: See the table above, it is exactly like the original dataframe, having 0 as a header. So, at this step, I should replace 0 header with my new_header I have created in Step 3 above

- Hence the below function will assign the **new_header** to **df_dep_t_new** dataframe.

```python
# Telling panda to take the values stored in the new_header variable and put them in the df_dep_t_new dataframe as the names of its columns.
df_dep_t_new.columns = new_header
```

## Step 6: Printing the final Trasposed Dataframe!

```python
# Printing out df_df_dep_new to check if my new_header is integrated in the df_dep_t_new dataframe
df_dep_t_new
```
